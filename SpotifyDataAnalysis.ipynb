```python
# 11. Clustering songs by audio features (KMeans example)
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = df_final[features].dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df_final.loc[X.index, 'cluster'] = clusters

plt.figure(figsize=(10,6))
sns.scatterplot(x='danceability', y='energy', hue='cluster', data=df_final, palette='Set2')
plt.title('KMeans Clustering of Songs by Danceability and Energy')
plt.show()
```

## Song suggestion algorithm

```python
df.columns
```

```python
features = ['year', 'danceability', 'energy', 'followers']
```

```python
X = df[features].copy()
```

```python
X = X.fillna(0)
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the features
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(X)
```

#### Train KNN model

```python
knn = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn.fit(X_scaled)
```

```python
df.describe()
```

```python
print("Please provide the following inputs:")

year= int(input("Release Year (e.g., 1900 to 2025): "))
danceability = float(input("Danceability (0.0 to 1.0): "))
energy = float(input("Energy (0.0 to 1.0): "))
followers = int(input("Artist Followers (e.g., 1000 to 100000000): "))
```

```python
user_input = [[year, danceability, energy, followers]]
```

```python
# Standardize user input
user_input_scaled = scaler.transform(user_input)
```

```python
# Find nearest neighbors
distances, indices = knn.kneighbors(user_input_scaled)
```

```python
# Fetch top 5 recommendations
print("\nüéµ Top 5 Recommended Songs for You:\n")

for idx in indices[0]:
    song_name = df.iloc[idx]['name']
    artist_name = df.iloc[idx]['artists']
    popularity = df.iloc[idx]['popularity']
    artist_followers = df.iloc[idx]['followers']
    Release_year=df.iloc[idx]['year']
    dancebility=df.iloc[idx]['danceability']
    energy=df.iloc[idx]['energy']
    #Release_year=df.iloc[idx]['year']
    
    print(f"Song: {song_name}")
    print(f"Artist: {artist_name}")
    print(f"Popularity: {popularity}")
    print(f"Followers: {artist_followers}")
    print(f"Year: {Release_year}")
    print(f"Danceability: {dancebility}")
    print(f"energy: {energy}")
    print("-" * 40)
```
